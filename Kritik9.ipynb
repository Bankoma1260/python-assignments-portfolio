```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

#a)
# Define the symbols and function
x, y = sp.symbols('x y')
f = sp.exp(x) * sp.sin(y) + y**3

# Compute partial derivatives
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("∂f/∂x =", df_dx)
print("∂f/∂y =", df_dy)

#b)

# Define the symbols and function
x, y = sp.symbols('x y')
g = x**2 * y + x * y**2

# Compute gradient vector components
grad_g_x = sp.diff(g, x)
grad_g_y = sp.diff(g, y)

# Gradient vector
grad_g = [grad_g_x, grad_g_y]

# Compute magnitude at point (1, -1)
point = {x: 1, y: -1}
magnitude = sp.sqrt(grad_g_x.subs(point)**2 + grad_g_y.subs(point)**2)

print("Gradient vector:", grad_g)
print("Magnitude at (1, -1):", magnitude)

#c)

# Define the symbols and function
x, y = sp.symbols('x y')
h = sp.ln(x**2 + y**2)

# Compute second partial derivatives
d2h_dx2 = sp.diff(h, x, x)
d2h_dy2 = sp.diff(h, y, y)
d2h_dxdy = sp.diff(h, x, y)

print("∂²h/∂x² =", d2h_dx2)
print("∂²h/∂y² =", d2h_dy2)
print("∂²h/∂x∂y =", d2h_dxdy)

# Discuss symmetry
print("Mixed partial derivatives are symmetric:", d2h_dxdy == sp.diff(h, y, x))

#d)

# Define the symbols and function
x, y = sp.symbols('x y')
k = sp.sin(x) * sp.cos(y)

# Convert to numerical function
k_func = sp.lambdify((x, y), k, 'numpy')

# Create grid and evaluate k
x_vals = np.linspace(-10, 10, 400)
y_vals = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = k_func(X, Y)

# Plot contour
plt.contourf(X, Y, Z, levels=50, cmap='plasma')
plt.colorbar()
plt.title('Contour plot of $k(x, y) = \sin(x) \cdot \cos(y)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

#e)

# Define the symbols and function
x, y = sp.symbols('x y')
p = x**2 - y**2

# Convert to numerical function
p_func = sp.lambdify((x, y), p, 'numpy')

# Create grid and evaluate p
x_vals = np.linspace(-5, 5, 400)
y_vals = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = p_func(X, Y)

# Plot contour
plt.contourf(X, Y, Z, levels=50, cmap='coolwarm')
plt.colorbar()
plt.title('Contour plot of $p(x, y) = x^2 - y^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

```
